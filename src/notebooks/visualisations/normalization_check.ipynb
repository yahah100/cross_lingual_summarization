# Normalization for better heatmap visualizations

## Idea
Use a L2 normalization and cut of outlieres wich would cause the heatmap to not showing small differences

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x = np.random.normal(1,2,(5000,2))
x = np.concatenate([x, np.random.normal(4,2,(400,2))])
x = np.concatenate([x, np.random.normal(8,2,(400,2))])
```

```python
def normalize(x, cut_outliers=True):
    # cut of outliers
    if cut_outliers:
        variance = np.var(x)
        mean = np.mean(x)
        x[x>(variance + mean)] = mean + variance
        x[x<(mean - variance)] = mean - variance
    # normalize
    x_res = x/np.linalg.norm(x)
    return x_res
```

```python
plt.scatter(x[:,0], x[:,1])
plt.show()
```

```python
x_n = normalize(x, False)
plt.scatter(x_n[:,0], x_n[:,1])
plt.show()
```

```python
x_n = normalize(x, True)
plt.scatter(x_n[:,0], x_n[:,1])
plt.show()
```
